-Obtener un json para **cada estaci√≥n meteorol√≥gica** donde se guarden los datos de
tr√°fico del mes de marzo de cada espira electromagn√©tica cuyas coordenadas se
encuentren dentro de un determinado radio a partir de las coordenadas de cada
estaci√≥n meteorol√≥gica. <br> Para ello se deben desagrupar todos los documentos de la
colecci√≥n de tr√°fico del mes de marzo por las parejas de coordenadas que tiene cada
tramo y guardar el resultado en una nueva colecci√≥n, generar un √≠ndice ‚Äú2dsphere‚Äù en
esta nueva colecci√≥n y utilizar la funci√≥n $geoNear del m√©todo $aggregate para realizar
la consulta geoespacial. Como radio de trabajo se puede poner una distancia de 600
metros desde cada estaci√≥n meteorol√≥gica. Por √∫ltimo, se deber√° exportar cada una de
las colecciones generadas en los correspondientes ficheros json.

```python
resume_data = database.agg("trafico", pipeline=[
    {
        "$match": {"$expr": {"$eq": [{"$month": "$Fecha"}, 3]}}
    },
    {
        "$group": {
            "_id":{ "month": {"$month":"$Fecha"},"idtramo": "$idtramo" },
            "avg_lectura":{"$avg": "$lectura"}
        }
    },
    {
        "$sort" : { "_id" : 1} 
    }
])
resume_data.head()
```
